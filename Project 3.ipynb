<h7> Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492 <h7>

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

<h7>We'll do anomaly detection using isolation forest and feature engineering to improve the data quality<h7>

```python
model = IsolationForest(n_estimators=100, contamination=0.2, max_features=1.0)
model.fit(new_dataset[['Amount']])
```

```python
new_dataset['anomalies_sccores']=model.decision_function(new_dataset[['Amount']])
new_dataset['anomaly']=model.predict(new_dataset[['Amount']])
new_dataset.head(20)
```

```python
#Number of Anamoalies
# Count the number of -1 values in the 'anomaly' column
num_anomalies = (new_dataset['anomaly'] == -1).sum()

print("Number of anomalies:", num_anomalies)
```

```python
# Remove entries with anomaly label -1
cleaned_dataset = new_dataset[new_dataset['anomaly'] != -1]

# Display the cleaned dataset
print(cleaned_dataset)
```

```python
cleaned_dataset.shape
```

<h7>Splitting the data into Features & Targets<h7>

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

<h7>Split the data into Training data & Testing Data<h7>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
